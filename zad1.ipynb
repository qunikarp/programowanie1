## Zadanie 15
Napisz prosty kalkulator.
Pobierz dwie wartości od użytkownika oraz rodzaj operacji (+,-,*,/) a następnie wyświetl na ekranie rozwiązanie.

przykład:
```shell
Podaj pierwszą liczbę > 1
Podaj drugą liczbę > 3
Podaj operację (+,-,*,/) > +
wynik: 4
```

````python
print("Dostępna lista operatorów dla kalkulatora to: +, -, *, /")
a=int(input("Podaj pierwszą liczbę: "))
b=int(input("Podaj drugą liczbę: "))
operacja = input("Rodzaj operacji: ")
if operacja=="+":
    print(a+b)
elif operacja=="-":
    print(a-b)
elif operacja=="*":
    print(a*b)
elif operacja=="/":
    print(a/b)
else: 
    print("Wprowadź prawidłowy znak!")
````

## Zadanie 16
Zmodyfikuj kalkulator z poprzedniego zadania tak, aby można było podawać liczby zmiennoprzecinkowe (`float`).

przykład:
```shell
Podaj pierwszą liczbę > 1.5
Podaj drugą liczbę > 3.5
Podaj operację (+,-,*,/) > -
wynik: -2.0
```

````python
print("Dostępna lista operatorów dla kalkulatora to: +, -, *, /")
print("Kalkulator automatycznie zaokrągla do dwóch miejsc po przecinku.")
a=float(input("Podaj pierwszą liczbę: "))
b=float(input("Podaj drugą liczbę: "))
operacja = input("Rodzaj operacji: ")
if operacja=="+":
    print(round((a+b),2))
elif operacja=="-":
    print(round((a-b),2))
elif operacja=="*":
    print(round((a*b),2))
elif operacja=="/":
    print(round((a/b),2))
else: 
    print("Wprowadź prawidłowy znak!")
````
